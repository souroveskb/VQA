```python
def parse_dataset(data):
  """Parses the given dataset into a dictionary.

  Args:
    data: A list of strings, where each string is in the format "image_name#instance_id    description".

  Returns:
    A dictionary where keys are image names and values are lists of descriptions for that image.
  """

  parsed_data = {}
  for line in data:
    image_name, description = line.split("    ", 1)  # Split into key and value
    image_name = image_name.split("#")[0]  # Remove instance ID if present

    if image_name not in parsed_data:
      parsed_data[image_name] = []  # Create a new list for the image
    parsed_data[image_name].append(description)

  return parsed_data

# Example usage:
dataset = [
    "1.jpg#0    একটি পার্কে একটি লেক আছে ।",
    "1.jpg#0    লেকের উপর একটি ব্রীজ আছে ।",
    # ... rest of your data
]
parsed_dict = parse_dataset(dataset)
print(parsed_dict)
```
